```python
from skimage.io import imread
from skimage.transform import resize
import numpy as np
import matplotlib.pyplot as plt
import pickle as pkl
import joblib
import os
import tensorflow as tf
import tensorflow_hub as hub  
from PIL import Image
```

```python
model_path = "/home/sacko/Documents/Projet_Stanford_Dogs2/Scripts/InceptionV3.h5"
model = tf.keras.models.load_model(model_path, custom_objects={'KerasLayer': hub.KerasLayer})
```

```python
CATEGORIES = ['n02085620-Chihuahua', 'n02085782-Japanese_spaniel',
       'n02085936-Maltese_dog', 'n02086079-Pekinese',
       'n02086240-Shih-Tzu', 'n02086646-Blenheim_spaniel']
```

```python
# Teste d'une sur nouvelle image
flat_data  = []
url = input("Entrer URL de votre image pour la pr√©diction")
img = imread(url)
img_resized = resize(img, (299, 299,3))
flat_data.append(img_resized.flatten())
falt_data = np.array(flat_data)
print(img.shape)
plt.imshow(img_resized)
```

```python
img_resize = tf.expand_dims(img_resized, 0)
predictions = model.predict(img_resize)

predicted_class_index = np.argmax(predictions)

y_out = CATEGORIES[predicted_class_index]
print(f"PREDICTION:{y_out}")
```
